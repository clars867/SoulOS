## Bose-Einstein Condensate

The fundamental equation used to describe the macroscopic, collective behavior of a BEC is the **Gross-Pitaevskii Equation (GPE)**. It's a non-linear Schrödinger equation for the condensate wavefunction.

### Gross-Pitaevskii Equation (GPE) 
for a BECThe time-dependent Gross-Pitaevskii Equation describes the evolution of the condensate wavefunction $\Psi(\mathbf{r}, t)$ at $\mathbf{r}$ and time $t$.

### Time-dependent Gross-Pitaevskii Equation

$$
i\hbar \frac{\partial \Psi(\mathbf{r}, t)}{\partial t} = \left[ -\frac{\hbar^2}{2m} \nabla^2 + V_{\text{ext}}(\mathbf{r}) + g|\Psi(\mathbf{r}, t)|^2 \right] \Psi(\mathbf{r}, t)
$$

#### Description of Terms

$i\hbar \frac{\partial \Psi}{\partial t}$ **(Left-Hand Side)**: This represents the temporal evolution of the wavefunction, characteristic of the Schrödinger equation.
    * $i$: The imaginary unit.
    * $\hbar$: The reduced Planck constant $(\frac{h}{2\pi})$.
    * $\Psi(\mathbf{r}, t)$: The condensate wavefunction or order parameter. Its magnitude squared, $|\Psi(\mathbf{r}, t)|^2$, gives the number density of the atoms in the condensate.

$-\frac{\hbar^2}{2m} \nabla^2$ **(Kinetic Energy Term)**: This is the kinetic energy operator.
    * $m$: The mass of a single atom.
    * $\nabla^2$: The Laplace operator (del squared).

$V_{\text{ext}}(\mathbf{r})$ **(External Potential)**: This is the potential energy from the external trap (like a magnetic or optical trap) confining the atoms. For a typical harmonic trap, $V_{\text{ext}}(\mathbf{r}) = \frac{1}{2}m(\omega_x^2 x^2 + \omega_y^2 y^2 + \omega_z^2 z^2)$, where $\omega_i$ are the trap frequencies.

$g|\Psi(\mathbf{r}, t)|^2$ **(Interaction Term)**: This non-linear term accounts for the atom-atom interactions within the condensate, which is what distinguishes the GPE from the standard linear Schrödinger equation.
    * $g$: The interaction strength parameter, defined as $g = \frac{4\pi \hbar^2 a_s}{m}$.
    * $a_s$: The s-wave scattering length, which determines if the interactions are repulsive $(a_s > 0)$ or attractive $(a_s < 0)$.

The external potential $V_{\text{ext}}$ could represent the physical structure or addressing scheme of the storage medium, and the non-linear interaction term $g|\Psi|^2$ could represent the collective behavior, stability, or error-correction mechanism of the stored data elements.

## Time-independent version of the Gross-Pitaevskii Equation (GPE)

This form is used to find the stationary states of the condensate, most importantly the ground state (the lowest energy state), by treating the energy E as an eigenvalue.

The time-independent GPE is derived by assuming a stationary state solution for the wavefunction of the form $\Psi(\mathbf{r}, t) = \phi(\mathbf{r}) e^{-i \mu t / \hbar}$, where $\phi(\mathbf{r})$ is the time-independent wavefunction and $\mu$ is the chemical potential (which is related to the ground state energy per particle).

$\left[ -\frac{\hbar^2}{2m} \nabla^2 + V_{\text{ext}}(\mathbf{r}) + g|\phi(\mathbf{r})|^2 \right] \phi(\mathbf{r}) = \mu \phi(\mathbf{r})$

### Description of Terms
This equation is an eigenvalue equation where the operator on the left, known as the Hamiltonian $\hat{H}_{\text{GPE}}$, acts on the time-independent wavefunction $\phi(\mathbf{r})$ to return the chemical potential $\mu$ multiplied by the same wavefunction.

  $\phi(\mathbf{r})$: The time-independent wavefunction of the condensate. $|\phi(\mathbf{r})|^2$ is the density of the condensate atoms at position $\mathbf{r}$.
  
  $\mu$ (Chemical Potential): This is the eigenvalue in the equation, representing the energy required to add one atom to the condensate without doing any work against the external potential. For the ground state, it's often close to the minimum energy per particle.
  **Other terms** $(-\frac{\hbar^2}{2m} \nabla^2$, $V_{\text{ext}}(\mathbf{r})$, and $g|\phi(\mathbf{r})|^2)$: These are the same Kinetic Energy, External Potential, and Interaction Energy terms as in the time-dependent GPE.
