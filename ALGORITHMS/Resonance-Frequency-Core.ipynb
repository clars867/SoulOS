## Resonance Frequency

### Ideal LC resonator

For an ideal inductor L and capacitor C, the natural angular resonant frequency is

$$
\omega_0=\frac{1}{\sqrt{LC}}
$$

and the resonant frequency in hertz is

$$
f_0=\frac{\omega_0}{2\pi}=\frac{1}{2\pi\sqrt{LC}}.
$$

### Realistic RLC (includes losses)

If you include resistance R, the resonance becomes a damped oscillator.

Series RLC
$$
\omega_0=\frac{1}{\sqrt{LC}},\qquad
\alpha=\frac{R}{2L},\qquad
\omega_d=\sqrt{\omega_0^2-\alpha^2}.
$$

Parallel RLC
A common form is
$$
\omega_0=\frac{1}{\sqrt{LC}},\qquad
\alpha=\frac{1}{2RC},\qquad
\omega_d=\sqrt{\omega_0^2-\alpha^2}.
$$

Where $\omega_d$ is the damped oscillation frequency.

### Quality factor and bandwidth

1) Series RLC:
$$
Q=\frac{\omega_0 L}{R}=\frac{1}{R}\sqrt{\frac{L}{C}}.
$$

2) Bandwidth (approximate, for reasonably high Q):
$$
\Delta f \approx \frac{f_0}{Q}.
$$
