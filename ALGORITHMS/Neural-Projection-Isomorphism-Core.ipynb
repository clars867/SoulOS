## Notes 1
**Translational Kinetic Energy ($\text{KE}$ or $K$)** is:

$$KE = \frac{1}{2} m v^2$$

Where:
* $m$ is the **mass** of the object (in kilograms, kg).
* $v$ is the **velocity** (speed) of the object (in meters per second, m/s).

This kinetic energy is the energy an object possesses due to its motion, and it is typically measured in **Joules (J)**.

The most common and fundamental equation for **average velocity ($\bar{v}$) ** is:

$$\bar{v} = \frac{\Delta d}{\Delta t}$$

Where:
* $\bar{v}$ is the **average velocity** (a vector quantity, including both speed and direction).
* $\Delta d$ (or $\Delta x$) is the **displacement** (the change in position, also a vector quantity).
* $\Delta t$ is the **time interval** over which the displacement occurred.

Since velocity is a **vector** quantity, you must always specify the **direction** of motion (e.g., $10\text{ m/s}$ north, or $-5\text{ m/s}$ in the x-direction).

### Other Key Velocity Equations

The equation you use depends on the situation, particularly whether the object is accelerating (changing its velocity) or moving at a constant velocity.

#### 1. Constant Acceleration (Kinematic Equations)

If an object is moving with a constant acceleration ($\text{a}$), you can find its **final velocity ($\text{v}$)** using this kinematic equation:

$$\text{v} = \text{u} + \text{at}$$

* $\text{v}$ is the **final velocity**.
* $\text{u}$ is the **initial velocity**.
* $\text{a}$ is the **acceleration** (constant).
* $\text{t}$ is the **time interval**.

#### 2. Instantaneous Velocity

In calculus-based physics, the velocity at a specific moment in time (the **instantaneous velocity**), $\text{v}(t)$, is the derivative of the object's position function, $x(t)$, with respect to time:

$$\text{v}(t) = \frac{d\text{x}}{d\text{t}}$$

## Notes 2 (Projector Physics)
The physics of projectors is based on the principles of geometric optics and involves mathematical formulas to determine image size, distance, and focus. 

### Thin Lens Equation 
The primary mathematical model used to describe how a projector lens forms an image is the thin lens equation. This equation relates the focal length of the lens to the distances of the image and the object (the internal display chip). 
$$
\frac{1}{f}=\frac{1}{d_{o}}+\frac{1}{d_{i}}
$$
Where: 
* $f$ is the focal length of the lens (a property of the lens itself). For the converging lenses used in projectors, the focal length is a positive value.
* $d_{o}$ is the object distance, the distance from the center of the lens to the image source (e.g., the DLP chip or LCD panel).* $d_{i}$ is the image distance, the distance from the center of the lens to the screen or projection surface. 

### Magnification and Image Size 
The equation for magnification $(M)$ determines how much larger the projected image is compared to the internal display chip. Magnification is a ratio of heights or distances. 
$$
M=\frac{h_{i}}{h_{o}}=-\frac{d_{i}}{d_{o}}
$$
Where: 
* $h_{i}$ is the height of the projected image.
* $h_{o}$ is the height of the original image source (the display chip).
* $d_{i}$ is the image distance.
* $d_{o}$ is the object distance. 

The negative sign in the formula for lenses indicates that the image produced is a real, inverted image. Projectors compensate for this inversion internally so the final image appears upright to the viewer. 

### Throw Ratio 
A practical application of the mathematics of projection is the throw ratio. This is a specific geometric relationship (similar triangles) that helps installers determine where to place a projector for a desired screen size. 
$$
\text{Throw\ Ratio}=\frac{\text{Throw\ Distance\ (D)}}{\text{Image\ Width\ (W)}}
$$ 
For example, a throw ratio of 2:1 means that for every 2 feet of distance from the screen, the image will be 1 foot wide.

### Digital Image Processing 
Beyond optics, the digital processing involves complex mathematical algorithms: Linear algebra is used to scale, rotate, and transform the pixel matrices (the image data) for various adjustments like keystone correction, which uses trigonometric calculations to make a trapezoidal image look rectangular when the projector is tilted. Calculus models light intensity to ensure even illumination and brightness across the entire screen.

## Notes 3
Vision will be represented “the way the brain reconstructs it” using a generative approach, and then invert it into a projection. The key is that mathematically, both are the same kind of system: a forward model plus an inverse (inference).
	
1) Brain-style reconstruction: the math view A clean abstraction is:
	1.1 World state $z$: objects, lighting, surfaces, motion
	1.2 Image $x$: the pixels on the retina (or in a camera)

Forward (rendering) model
	$x = f(z) + \epsilon$

This says: given a scene $z$, physics + optics produce an image $x$, plus noise $\epsilon$.

2) Inverse (perception) problem
$$
\hat z = \arg\max_z \; p(z \mid x)
$$

Or equivalently by Bayes:
* $p(z \mid x) \propto p(x \mid z)\,p(z)$
* $p(x \mid z)$ is “how likely this image is if the world were $z$” (your optics + sensor model)
* $p(z)$ is a prior [prior: an expectation/bias before seeing data], like “light usually comes from above,” “edges tend to continue,” etc.

This framing matches predictive coding: brain-as-prediction where perception minimizes prediction error, where the system constantly predicts $x$ from $\tilde z$, compares to actual $x$, and updates $\tilde z$ to reduce error.

So “brain reconstructs vision” can be written as minimizing an energy functional [a scalar objective defined over functions/fields]:
$$
E(z)=\underbrace{\|x - f(z)\|^2}_{\text{prediction error}}+\lambda \underbrace{R(z)}_{\text{regularizer/prior}}
$$
where $R(z)$ encodes your constraints (smoothness, sparsity, edges, etc.).

2) Inverting it into a “projector” hypothesis

If you flip the story: instead of “world --> to retina --> to brain,” you imagine the brain generates an internal image [or sturctural map] $\tilde x = f(\tilde z)$ and that becomes your experienced world.

Mathematically, that’s just:
$$
\tilde x = f(\tilde z)
$$
and perception becomes a fusion of:
* bottom-up signal (actual sensory $x$)
* top-down prediction $\tilde x$

A simple fusion equation looks like:
$$
x_{\text{perceived}} = \alpha x + (1-\alpha)\tilde x
$$
where $$\alpha$$ changes with attention, uncertainty, fatigue, etc.

*Important nuance: the projector concept is used to describe how the brain projects a model onto ambiguous input (which is basically what strong priors do). That’s why illusions exist: when $p(z)$ dominates $p(x\mid z)$, the “projection” wins.*

3) Simpler Equations: Optical Projections

A projector is basically an “inverse camera.” Using homogeneous coordinates:
$$
\mathbf{u} \sim \mathbf{K} [\mathbf{R}\mid \mathbf{t}] \mathbf{X}
$$
* $\mathbf{X}$ is a 3D point in the world
* $\mathbf{u}$ is a pixel location on the projector (or camera) image plane
* $\mathbf{K}$ is intrinsics (focal length, principal point)
* $\mathbf{R}$, $\mathbf{t}$ are extrinsics (pose)

If the surface is planar, it collapses to a homography.
$$
\mathbf{u} \sim \mathbf{H}\mathbf{x}
$$

4) Photometry (how bright it looks)

A minimal brightness model:
$$
I_{\text{seen}}(p)= \rho(p)\, \big(I_{\text{proj}}(u(p))\big)^\gamma + \eta
$$
* $\rho(p)$ surface reflectance (albedo)
* $\gamma$ gamma curve (projector nonlinearity)
* $\eta$$ ambient + noise

5) Blur / diffraction (wave-ish part)

Projectors and eyes both have a point spread function:
$$
I_{\text{seen}} = I_{\text{ideal}}*h
$$
Convolution with $h$ is where the “image as waveform” becomes concrete: $h$ is an impulse response, just like in signal processing.

The operational optics gives you a crisp chain of transforms:
* warp (geometry) → nonlinearity (gamma) → convolution (PSF) → sampling.


6) Alternative hybrid accounting

One elegant hybrid:
$$
\tilde x_{k+1} = \tilde x_k + \mathbf{D}\,\nabla_{\tilde x}\Big(\text{match}(x,\tilde x_k)\Big)
$$
where $\mathbf{D}$ is the decay/attenuation matrix that controls how aggressively the internal reconstruction updates.

Isomorphicic Projection:
 The brain-as-inference and projector-as-forward-model are almost isomorphic as frameworks; they just swap what’s known vs. what’s solved for.
