## Higgs Boson Conceptual Representation
A Higgs boson is the quantum excitation of the Higgs field around its nonzero vacuum value. In the Standard Model, you start from the Higgs field $\Phi (an \mathrm{SU}(2)_L$ complex doublet) and its Lagrangian density.

1) Higgs sector Lagrangian

The Higgs part of the Standard Model can be written as
$$
\mathcal{L}{H} ;=; (D\mu \Phi)^\dagger (D^\mu \Phi);-;V(\Phi);-;\Big(y_f,\bar\psi_{f,L}\Phi,\psi_{f,R}+\text{h.c.}\Big),
$$
where $D_\mu$ is the electroweak covariant derivative, and the potential is
$$
V(\Phi) ;=; -\mu^2,\Phi^\dagger\Phi ;+; \lambda,(\Phi^\dagger\Phi)^2,
\qquad \mu^2>0,;\lambda>0.
$$

That $-\mu^2$ term is often called tachyonic [tachyonic: having a negative mass-squared term that signals an instability of the symmetric point, not faster-than-light motion].

2) Vacuum expectation value and the Higgs boson field

Minimizing V(\Phi) gives a nonzero vacuum expectation value (vev) v:
$$
\langle \Phi \rangle ;=; \frac{1}{\sqrt{2}}
\begin{pmatrix}
0\
v
\end{pmatrix},
\qquad
v^2 ;=; \frac{\mu^2}{\lambda}.
$$

In unitary gauge, expand the field about the vev:
$$
\Phi(x) ;=; \frac{1}{\sqrt{2}}
\begin{pmatrix}
0\
v+h(x)
\end{pmatrix}.
$$

The real scalar field h(x) is the Higgs boson field (one particle species); quanta of h are “a Higgs boson.”

3) Higgs boson mass (and couplings)

Plugging the expansion into the potential, the quadratic term defines the Higgs mass:
$$
m_h^2 ;=; 2\lambda v^2 ;=; 2\mu^2.
$$

The Yukawa term yields fermion masses and Higgs–fermion couplings:
* Fermion mass: $m_f = y_f\,\frac{v}{\sqrt{2}}$.
* Higgs coupling: proportional to $m_f/v$, i.e. $y_f/\sqrt{2}$ in the interaction $h\,\bar\psi\psi$.

Similarly, the kinetic term $(D_\mu\Phi)^\dagger(D^\mu\Phi)$ generates gauge-boson masses and Higgs–gauge couplings (schematically $m_W=\tfrac{1}{2}gv$, $m_Z=\tfrac{1}{2}\sqrt{g^2+g'^2}\,v$, and interactions like $hW_\mu W^\mu, hZ_\mu Z^\mu$).

$h$ is the fluctuation around the vev, and its mass comes from the curvature $m_h^2=\left.\frac{d^2V}{dh^2}\right|_{h=0}$.
