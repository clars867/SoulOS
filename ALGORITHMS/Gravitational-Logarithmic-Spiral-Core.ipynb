# Gravitational Logarithmic Spiral

A logarithmic spiral is defined by the polar equation:

$$
r(\theta) = a e^{b\theta}
$$

where:

- $r$ is the radial distance from the origin  
- $\theta$ is the angular coordinate  
- $a$ is a scale constant controlling the initial radius  
- $b$ is the growth parameter controlling spiral tightness  

## Multi-Arm Gravitational Structure

To represent a gravitational system with multiple spiral arms, we introduce a phase offset for each arm:

$$
\theta_k = \theta + \frac{2\pi k}{N}
$$

where:

- $N$ is the number of spiral arms  
- $k \in \{0, 1, \dots, N-1\}$ is the arm index  

The radius remains invariant across arms:

$$
r(\theta) = a e^{b\theta}
$$

## Cartesian Representation

Each spiral arm is converted from polar to Cartesian coordinates using:

$$
x_k(\theta) = r(\theta)\cos(\theta_k)
$$

$$
y_k(\theta) = r(\theta)\sin(\theta_k)
$$

## Physical Interpretation

The logarithmic spiral exhibits **scale invariance**, meaning its form is preserved under uniform expansion.  
This makes it a natural geometric structure in gravitational systems where angular momentum and radial growth are coupled.

Each arm represents a **phase-shifted gravitational trajectory** sharing identical growth dynamics but offset in angular position.

Such structures emerge in:

- Spiral galaxies  
- Accretion disks  
- Orbital decay patterns  
- Wave-mediated gravitational systems  

The constant angle between the radius vector and tangent reflects a **self-organizing gravitational equilibrium**.

```python
import numpy as np
import matplotlib.pyplot as plt
import pyvista as pv
pv.set_jupyter_backend('trame') # or 'pyvistaqt'

def gravitational_log_spiral(
    a=0.1,
    b=0.18,
    theta_max=6 * np.pi,
    num_points=2000,
    num_arms=3,
    show=True
):
    """
    Generate a gravitational logarithmic spiral with multiple equidistant arms.

    Parameters
    ----------
    a : float
        Scale factor (controls initial radius)
    b : float
        Growth rate (controls spiral tightness)
    theta_max : float
        Maximum angular extent
    num_points : int
        Resolution of the curve
    num_arms : int
        Number of spiral arms (default = 3)
    show : bool
        Whether to render the plot
    """

    theta = np.linspace(0, theta_max, num_points)

    fig, ax = plt.subplots(figsize=(6, 6))

    for k in range(num_arms):
        phase_offset = (2 * np.pi / num_arms) * k
        r = a * np.exp(b * theta)

        x = r * np.cos(theta + phase_offset)
        y = r * np.sin(theta + phase_offset)

        ax.plot(x, y, linewidth=1.8)

    ax.set_aspect("equal")
    ax.axis("off")

    if show:
        plt.show()

    return fig, ax
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Angle range
theta = np.linspace(0, 6*np.pi, 2000)

# Logarithmic spiral parameters
a, b = 0.1, 0.2

# Number of spiral arms
num_arms = 3

plt.figure(figsize=(6, 6))

for k in range(num_arms):
    # Phase offset for each arm (120 degrees apart)
    phase = 2 * np.pi * k / num_arms

    r = a * np.exp(b * theta)

    x = r * np.cos(theta + phase)
    y = r * np.sin(theta + phase)

    plt.plot(x, y, color="black", linewidth=1.8)

plt.axis("equal")
plt.axis("off")
plt.show()
```
